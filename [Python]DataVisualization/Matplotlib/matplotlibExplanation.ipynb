# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# To show the plots in the notebook. If you are using another IDE, you'll need to type 'plt.show()' at the end of your code.
%matplotlib inline
```

```python
# The data
x = np.linspace(0, 5, 11)
y = x ** 2
```

## Matplotlib Object Oriented Method

This means we will instantiate figure objects and then call methods or attributes from that object.

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y, 'b')
axes.set_xlabel("Set Xlabel")
axes.set_ylabel("Set Ylabel")
axes.set_title("Title")
```

## Subplots

This subplots creates a frame where you can place the graphs.

```python
fig, axes = plt.subplots()

axes.plot(x, y, 'b')
axes.set_xlabel("Set Xlabel")
axes.set_ylabel("Set Ylabel")
axes.set_title("Title")
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x, y, 'b')
axes[0].set_xlabel("Set Xlabel")
axes[0].set_ylabel("Set Ylabel")
axes[0].set_title("Title")

axes[1].plot(x, y, 'red')
axes[1].set_xlabel("Set Xlabel")
axes[1].set_ylabel("Set Ylabel")
axes[1].set_title("Title")
```

**fig.tight_layout()** or **plt.tight_layout()** adjusts the picture so that there is no overlap in it.

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x, y, 'b')
axes[0].set_xlabel("Set Xlabel")
axes[0].set_ylabel("Set Ylabel")
axes[0].set_title("Title")

axes[1].plot(x, y, 'red')
axes[1].set_xlabel("Set Xlabel")
axes[1].set_ylabel("Set Ylabel")
axes[1].set_title("Title")

fig.tight_layout()
```

### Figure size

```python
fig, axes = plt.subplots(figsize=(15,4), nrows = 1, ncols = 2)

axes[0].plot(x, y, 'b')
axes[0].set_xlabel("Set Xlabel")
axes[0].set_ylabel("Set Ylabel")
axes[0].set_title("Title")

axes[1].plot(x, y, 'red')
axes[1].set_xlabel("Set Xlabel")
axes[1].set_ylabel("Set Ylabel")
axes[1].set_title("Title")

fig.tight_layout()
```

### Saving figures

To save a figure to a file we can use the savefig method in the Figure class. We can specify the desired format of the image on the extension of the file. 

```python
fig.savefig("filename.png")
fig.savefig("filename.jpg")
```

### Legends

You can use the **label="label text"** keyword argument when plots or other objects are added to the figure, and then using the **legend** method without arguments to add the legend to the figure. 

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend()
```

We can also specify where the legend is going to be placed with the argumnet `loc`.

```python
ax.legend(loc=0) # let matplotlib decide the optimal location
fig
```

## Color, line type, marker and etc

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y,color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
axes.set_xlabel("Set Xlabel")
axes.set_ylabel("Set Ylabel")
axes.set_title("Title")
```
